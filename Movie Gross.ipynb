```python
%matplotlib inline 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import Imputer
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, r2_score
from sklearn.metrics import roc_curve, f1_score, classification_report
from math import sqrt

from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn import neighbors
```

### Data-sets

```python
# training datasets merged together (2000-2002, 2005-2007, 2010-2012)
# testing datasets merged together (2003,2008,2013)
# evaluation datasets merged together (2004,2009,2014)
movie_gross_training = pd.read_csv("movie_gross_training.csv")
movie_gross_testing = pd.read_csv("movie_gross_testing.csv")
movie_gross_evaluation = pd.read_csv("movie_gross_evaluation.csv")
del movie_gross_training['hi_lo']
del movie_gross_testing['hi_lo']
del movie_gross_evaluation['hi_lo']
```

### Pre-Processing

```python
# Converting Dataframes into row lists
# Converting rows into columns
# allRowsList = []
# for idx,row in movie_gross_training.iterrows():
#     currentRowList = row.values.tolist()
#     allRowsList.append(currentRowList)
training_gross_values = []
training_param_values = []
for idx,row in movie_gross_training.iterrows():
    rowEntry = row.values.tolist()
    grossValue = rowEntry[0]
    parameterList = rowEntry[1:]
    training_gross_values.append(grossValue)
    training_param_values.append(parameterList)

testing_gross_values = []
testing_param_values = []
for idx,row in movie_gross_testing.iterrows():
    rowEntry = row.values.tolist()
    grossValue = rowEntry[0]
    parameterList = rowEntry[1:]
    testing_gross_values.append(grossValue)
    testing_param_values.append(parameterList)
```

```python
# Filling in missing values with mean
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp.fit(training_param_values)
training_param_values = imp.transform(training_param_values)

imp.fit(testing_param_values)
testing_param_values = imp.transform(testing_param_values)
```

```python
col_predict_gross = 'predict_gross'
col_gross = 'gross'
col_error = 'error'
col_rel_error = 'relative_error'
```

### Testing the OLS Model on our test data set

```python
def evaluation(modelOutput):
    modelOutput[col_error] = (modelOutput[col_predict_gross]-modelOutput[col_gross])
    modelOutput[col_rel_error] = (modelOutput[col_predict_gross]-modelOutput[col_gross])*100/modelOutput[col_gross]
    medianRelError = modelOutput[col_rel_error].median()
    minRelError = modelOutput[col_rel_error].min()
    maxRelError = modelOutput[col_rel_error].max()
    
    rmse = sqrt(mean_squared_error(modelOutput[col_gross], modelOutput[col_predict_gross]))
    r2score = r2_score(modelOutput[col_gross], modelOutput[col_predict_gross])
    
    print 'Evaluation Result'
    print '=================================================================='
    print 'Total Number of Instances:', modelOutput.shape[0]
    print 'Median Relative Absolute Error:', medianRelError, '%'
    print 'Min Relative Absolute Error:', minRelError, '%'
    print 'Max Relative Absolute Error:', maxRelError, '%'
    print 'The root mean square error:' , rmse
    print 'The r2 score:', r2score
    
    # Absoluate Error Distribution
    binNum=400
    fig1, absErrorAxe = plt.subplots(1, 1, figsize=(18,9))
    n, bins, patches = absErrorAxe.hist(modelOutput[col_error], binNum, facecolor='blue', alpha=0.75, \
                                        linewidth=0, range=(-150000000, 150000000))
    absErrorAxe.set_title("Movie Gross Relative Absolute Error Distribution", fontsize=24, fontweight='bold')
    absErrorAxe.set_xlabel('Absolute Error', fontsize=22)
    absErrorAxe.set_ylabel('Number of Movies', fontsize=22)
    plt.show()
    
    return modelOutput, medianRelError, minRelError, maxRelError
    
```

###Baseline

```python
genres = movie_gross_training.columns[10:]
```

```python
def calculateGenreAverage():
    genreAverage={}
    for key in genres:
        genreAverage[key] = np.mean(movie_gross_training.loc[( movie_gross_training[key]==1),:]['gross'])
    return genreAverage
```

```python
genreAvg = calculateGenreAverage()
```

```python
genreAvg
```

```python
def calAverageofAverage():
    genredf = movie_gross_testing[genres]
    testOutput =[]
    for index, row in genredf.iterrows():
        count = 0 
        sumAvgs = 0
        for genreType in genres:
            if row[genreType]==1:
                count = count+1
                sumAvgs = sumAvgs + genreAvg[genreType]
        testOutput.append(sumAvgs/count*1.0)
    return testOutput
```

```python
baselineResults = calAverageofAverage()
```

```python
BaselineModel = pd.DataFrame(testing_gross_values)
BaselineModel.columns = [col_gross]
BaselineModel[col_predict_gross] = baselineResults
```

```python
modelOutput, medianRelError, minRelError, maxRelErrorevaluation = evaluation(BaselineModel)
```

### Split the movies
Split movies into high gross and low grossing movies

```python
# splitting training datasets
movie_gross_cutoff = np.percentile( movie_gross_training['gross'],25)
high_gross_movies_training = movie_gross_training.loc[( movie_gross_training['gross']>movie_gross_cutoff),:]
low_gross_movies_training = movie_gross_training.loc[(movie_gross_training['gross']<=movie_gross_cutoff),:]
```

```python
# splitting testing datasets
high_gross_movies_testing = movie_gross_testing.loc[( movie_gross_testing['gross']>movie_gross_cutoff),:]
low_gross_movies_testing = movie_gross_testing.loc[(movie_gross_testing['gross']<=movie_gross_cutoff),:]
```

```python
# splitting evaluation datasets
high_gross_movies_evaluation = movie_gross_evaluation.loc[( movie_gross_evaluation['gross']>movie_gross_cutoff),:]
low_gross_movies_evaluation = movie_gross_evaluation.loc[(movie_gross_evaluation['gross']<=movie_gross_cutoff),:]
```
